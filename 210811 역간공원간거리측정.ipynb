```python
from haversine import haversine
```

```python
import pandas as pd
df = pd.read_excel('데이터관련/성남판교/park.xlsx')
park = df[['공원명','위도','경도']]
park = park.values.tolist()
park
```

```python
df2 = pd.read_excel('데이터관련/성남판교/sungnamtrain.xlsx')
df2.head()
train = df2[['지하철','위도','경도']]
train = train.values.tolist()
train
```

```python
df3 = pd.read_excel('데이터관련/성남판교/sungnamprice2.xlsx')
df3.head()
apt = df3[['도로명','위도','경도']]
apt = apt.values.tolist()
apt
```

```python
dist_apt_train = []
for i in range(len(apt)):
    distan = []
    for j in range(len(train)):
        d = haversine(apt[i][1:3],train[j][1:3], unit = 'km')
        distan.append(d)
    dist_apt_train.append(distan)

apt_n = [i[0] for i in apt]
train_n = [i[0] for i in train]

data_df = pd.DataFrame(dist_apt_train, index=apt_n, columns=train_n)
data_df
```

```python
df_result = pd.concat([data_df.idxmin(axis=1),data_df.min(axis=1)],axis=1)
df4 = pd.DataFrame(df_result)
df4.columns=['역명', '역과의거리(km)']
df4
```

```python
dist_apt_park = []
for i in range(len(apt)):
    distan = []
    for j in range(len(park)):
        d = haversine(apt[i][1:3],park[j][1:3], unit = 'km')
        distan.append(d)
    dist_apt_park.append(distan)

apt_n = [i[0] for i in apt]
park_n = [i[0] for i in park]

data_df2 = pd.DataFrame(dist_apt_park, index=apt_n, columns=park_n)
data_df2.head()
```

```python
df2_result = pd.concat([data_df2.idxmin(axis=1),data_df2.min(axis=1)],axis=1)
df5 = pd.DataFrame(df2_result)
df5.columns=['공원명', '공원과의거리(km)']
df5
```

```python
result = pd.concat([df4, df5],axis=1)
result.to_excel(excel_writer='parktrain_distance.xlsx') 
```
