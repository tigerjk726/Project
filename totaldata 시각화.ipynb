```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

set(sorted([f.name for f in mpl.font_manager.fontManager.ttflist]))
mpl.rc('font', family = "NanumGothicCoding")
mpl.rc('axes', unicode_minus=False)

df = pd.read_excel('데이터관련/data_total.xlsx')
df
```

도시별 유해시설 및 기피시설 현황

```python
hate1 = df['정신보건개소'] *2
hate2 = df['양로시설&노숙시설']
hate3 = df['분뇨처리시설'] *10
hate4 = df['폐기물처리시설']/10
hate5 = df['화장시설&묘지']
city = df['도시명']
 

fig = plt.figure(figsize=(15,8)) 
ax = fig.add_subplot()

ax.plot(city, hate1, marker='o', label='정신보건개소') 
ax.plot(city, hate2, marker='o', label='양로시설&노숙시설')
ax.plot(city, hate3, marker='o', label='분뇨처리시설')
ax.plot(city, hate4, marker='o', label='폐기물처리시설')
ax.plot(city, hate5, marker='o', label='화장시설&묘지')

ax.legend() 
plt.xticks(rotation=45)
plt.show()
```

도시별 유해시설 및 기피시설 현황

```python
popul = df['인구수'] / 10000
price = df['실거래가'] / 1000
city = df['도시명']
 

fig = plt.figure(figsize=(15,8)) 
ax = fig.add_subplot()

ax.plot(city, popul, marker='x', label='인구수')
ax.plot(city, price, marker='s', label='실거래가')

ax.legend() 
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = df.corr()
df
```

```python
fig, ax = plt.subplots( figsize=(15,10) )

mask = np.zeros_like(df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(df, 
            annot = True,   
            mask=mask,     
            linewidths=.5,  
            vmin = -1,vmax = 1  
           )  
plt.show()
```
