```python
import pandas as pd
import numpy as np
import seaborn as sns
import folium
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('데이터관련/houseprice2021.xlsx')
df.head()
```

```python
df.info()
```

```python
df7085 = df['전용면적(㎡)'] >= 70

df = df[df7085]

df
```

```python
df_group = df.groupby(['시군구'], as_index=False).mean()
```

```python
df_group
```

```python
boondang = df_group['시군구'].str.contains("분당동|수내동|정자동|서현동|이매동|야탑동|분당구 금곡동|구미동")

boondang_df = df_group[boondang]

boondang_df
```

```python
pyeongchon = df_group['시군구'].str.contains("비산동|관양동|호계동|평촌동")

pyeongchon_df = df_group[pyeongchon]

pyeongchon_df
```

```python
joongdong = df_group['시군구'].str.contains("부천시 상동|부천시 중동|대산동")

joongdong_df = df_group[joongdong]

joongdong_df
```
