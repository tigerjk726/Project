```python
import pandas as pd
```

```python
apt = pd.read_csv('/content/drive/MyDrive/dataset/성남시 아파트 단지 - Sheet4 (1).csv')
apt = apt[['단지명', '위도', '경도','가격']]
apt
```

```python
park = pd.read_csv('/content/drive/MyDrive/dataset/성남시 아파트 단지 - Sheet2.csv')
park = park[['공원명', '위도', '경도']]
park
```

```python
train = pd.read_csv('/content/drive/MyDrive/dataset/성남시 아파트 단지 - Sheet1.csv')
train = train[['name', 'lat', 'lng']]
train
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/sungnam - pankyo.csv')
geo_data = '/content/drive/MyDrive/dataset/map (13).geojson'
```

```python
import folium

center = [37.378, 127.113]
m = folium.Map(location=center, tiles='cartodbpositron', zoom_start=13)

for i in apt.index[:]:
    folium.Circle(
        location = apt.loc[i, ['위도', '경도']],
        tooltip = apt.loc[i, '단지명'],
        color='black',
        fill=apt.loc[i, '가격'],
        radius = 20
    ).add_to(m)

for i in park.index[:15]:
    folium.Circle(
        location = park.loc[i, ['위도', '경도']],
        tooltip = park.loc[i, '공원명'],
        color='green',
        fill='True',
        radius = 500
    ).add_to(m)

for i in train.index[9:]:
    folium.Circle(
        location = train.loc[i, ['lat', 'lng']],
        tooltip = train.loc[i, 'name'],
        color='red',
        fill='True',
        radius = 750
    ).add_to(m)


# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.Choropleth(
    geo_data=geo_data,
    data=df,
    columns=('adm_cd','평균가격'),
    key_on='feature.properties.adm_cd',
    fill_color='PuBu',
    legend_name='아파트매매 평균가'
).add_to(m)

m
```
