```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

set(sorted([f.name for f in mpl.font_manager.fontManager.ttflist]))
mpl.rc('font', family = "NanumGothicCoding")
mpl.rc('axes', unicode_minus=False)

df = pd.read_excel('데이터관련/시군별주택평균가격.xlsx')
df
```

```python
area = df['시군구'] 
price = df['평균가격(만원)']
 
plt.figure(figsize=(10,10))
ytick_label_position = list(range(len(area)))
plt.yticks(ytick_label_position, area)
 
plt.barh(ytick_label_position, price) 
 
plt.title('시군별 주택평균가격',fontsize=20) 
plt.xlabel('평균가격(만원)') 
plt.ylabel('시군구')
plt.show()
```
